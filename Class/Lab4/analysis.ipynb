```python
# Using Colormaps
x = np.array([80,85,90,95,100,105,110,115,120,130,145])
y = np.array([240,250,260,270,300,310,320,330,350,360,400])

sizes = np.array([80,85,90,95,100,105,110,115,120,130,145])
color_map= np.array([20,30,40,50,60,70,80,90,67,89,27])

plt.scatter(x,y, c= color_map , cmap = 'viridis', s= sizes, alpha = 0.1 )
plt.show()
```

```python
from random import randint
x = []
y = []

for i in range(11):
    x.append(randint(0,100))
    
for i in range(11):
    y.append(randint(0,100))

color = []
size = []
for i in range(11):
    color.append(randint(0,100))

for i in range(11):
    size.append(randint(0,100))

plt.scatter(x,y, c= color , cmap = 'viridis', s= size)
plt.show()
```
