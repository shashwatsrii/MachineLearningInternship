```python
import pandas as pd
```

```python
df = pd.read_csv("Customer.csv")
```

```python
df.head()
```

```python
df_new = df.iloc[:,2:]
df_new
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df_new.iloc[:,0:2],df.iloc[:,2],test_size=0.2)
```

```python
print(X_test.shape)
print(X_train.shape)
print(Y_test.shape)
print(Y_train.shape)
```

```python
oe = OrdinalEncoder(categories = [['Poor','Average','Good'],['School','UG', 'PG']])
```

```python
oe.fit(X_train)
```

```python
X_train_new = oe.transform(X_train)
X_test_new = oe.transform(X_test)
```

```python
X_train_new
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(Y_train)
Y_train_new = le.transform(Y_train)
Y_test_new = le.transform(Y_test)
```
